```python
!kaggle datasets download -d omkargurav/face-mask-dataset

import zipfile
with zipfile.ZipFile('face-mask-dataset.zip', 'r') as zip_ref:
    zip_ref.extractall('mask_dataset')

print("Dataset downloaded and extracted!")
```

```python
import os
import shutil

os.makedirs("dataset/train", exist_ok=True)
os.makedirs("dataset/val", exist_ok=True)

classes = ["with_mask", "without_mask"]

for cls in classes:
    src = f"mask_dataset/data/{cls}"
    files = os.listdir(src)

    split = int(len(files) * 0.8)  # 80% train, 20% val

    train_dst = f"dataset/train/{cls}"
    val_dst = f"dataset/val/{cls}"

    os.makedirs(train_dst, exist_ok=True)
    os.makedirs(val_dst, exist_ok=True)

    for f in files[:split]:
        shutil.copy(os.path.join(src, f), train_dst)

    for f in files[split:]:
        shutil.copy(os.path.join(src, f), val_dst)

print("Dataset organized into train/val folders!")
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = 224
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(
    rescale=1./255
)

train_generator = train_datagen.flow_from_directory(
    "dataset/train",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary"
)

val_generator = val_datagen.flow_from_directory(
    "dataset/val",
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode="binary"
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

def create_model():
    base = tf.keras.applications.MobileNetV2(
        input_shape=(IMG_SIZE, IMG_SIZE, 3),
        include_top=False,
        weights="imagenet"
    )
    base.trainable = False  # Freeze pretrained model

    model = models.Sequential([
        base,
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation="relu"),
        layers.Dropout(0.3),
        layers.Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["accuracy"]
    )

    return model

model = create_model()
model.summary()
```

```python
EPOCHS = 5  # Fast training, 5–10 min

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=EPOCHS
)
```

```python
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(acc, label='Train Accuracy')
plt.plot(val_acc, label='Validation Accuracy')
plt.title("Accuracy")
plt.legend()

plt.subplot(1,2,2)
plt.plot(loss, label='Train Loss')
plt.plot(val_loss, label='Validation Loss')
plt.title("Loss")
plt.legend()

plt.show()
```

```python
import gradio as gr
import numpy as np

def predict_mask(img):
    img = tf.image.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img / 255.0
    img = tf.expand_dims(img, 0)

    pred = model.predict(img)[0][0]

    return {
        "With Mask": float(1 - pred),
        "Without Mask": float(pred)
    }

iface = gr.Interface(
    fn=predict_mask,
    inputs=gr.Image(),
    outputs=gr.Label(),
    title="Mask Detection CNN",
    description="Upload an image to detect if the person is wearing a mask."
)

iface.launch()
```
